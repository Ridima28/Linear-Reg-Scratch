Write a Python function that performs linear regression using the normal equation. The function should take a matrix X (features) and a vector y (target) as input, and return the coefficients of the linear regression model. Round your answer to four decimal places, -0.0 is a valid result for rounding a very small number.

```python
import numpy as np 

class LinearRegressionNorm:
    def __init__(self):
        self.theta = None 
        #theta is weight+bias

    def fit(self, X: list[list[float]], y :list[float]):
        X = np.array(X)
        y = np.array(y)
        self.theta = np.linalg.inv(X.T @ X ) @ X.T @ y  
        # inverse of (X transpose @ X) @ X transpose @ y) = Normal form equation 
        # "@" is matrix multiplication's sign 
        return self 
    
    def predict(self, X: list[list[float]]) -> np.ndarray:
        X = np.array(X)
        return np.round(X @ self.theta, 4)
```

TES CASE 1-> 

FEATURES (X) => [[1,1],[1,2],[1,3]]

TARGET COL (y) => [1,2,3]

```python
X = [[1,1],[1,2],[1,3]]
y = [1,2,3]
model = LinearRegressionNorm()

model.fit(X,y)
print("Weights given to each features: ", np.round(model.theta,4).tolist())
print("Predictions: ", model.predict([[1,1],[1,2],[1,3]]))
```
